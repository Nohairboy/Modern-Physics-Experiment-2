# 微波与铁磁共振实验

```python
import pandas as pd
path = r'C:\Users\86136\OneDrive\Desktop\2025近物实验\Cyh\微波和铁磁共振\铁磁共振data.xlsx'
df = pd.read_excel(path, sheet_name='Sheet1')

import matplotlib.pyplot as plt
from matplotlib import rcParams

# 设置中文字体和负号正常显示
rcParams['font.sans-serif'] = ['SimHei']
rcParams['axes.unicode_minus'] = False
```

```python
def get_data(x_col, y_col):
    
    x = df.values[:,x_col]
    x = x[~pd.isnull(x)]
    x = [x2 for x2 in x if isinstance(x2, (float, int))]
    print(x)

    y = df.values[:,y_col]
    y = y[~pd.isnull(y)]
    y = [val*1 for val in y if isinstance(val, (int, float))]
    print(y)

    return x, y

def my_plot(x, y, my_x_label, my_y_label, my_label='', my_title='', my_figsize=(10, 6)):
    plt.figure(figsize=my_figsize)
    plt.plot(x, y, marker='o', color='red', label=my_label)

    plt.title(my_title)
    plt.xlabel(my_x_label)
    plt.ylabel(my_y_label)

    plt.grid(True)
    plt.legend()
    plt.show()
```

## 1.2.1

```python
x2, y2 = get_data(x_col = 0, y_col = 1)
my_plot(x2, y2, my_x_label='$U$ (V)', my_y_label='$I$ (mA)', my_label='$I-U$ 曲线',
         my_title=r'体效应管在 0V-13V 的 $I-U$ 曲线', my_figsize=(10, 6))
```

## 2.7 用高斯计测量电磁铁电流与磁场强度的关系

```python
x8, y8 = get_data(x_col = 25, y_col = 26)

import numpy as np
from sklearn.linear_model import LinearRegression

X_train = np.array(x8).reshape(len(x8), 1)  
Y_train = np.array(y8).reshape(len(y8), 1)  

lineModel = LinearRegression()
lineModel.fit(X_train, Y_train)

Y_predict = lineModel.predict(X_train)

#coef_是系数，intercept_是截距
k = lineModel.coef_[0]
b = lineModel.intercept_[0]
print("y=%.4f*x+%.4f" % (k,b))

#对回归模型进行评分，这里简单使用训练集进行评分，实际很多时候用其他的测试集进行评分
print("R2 = ", lineModel.score(X_train, Y_train))


plt.figure(figsize=(10,6))
plt.scatter(x8, y8, marker='o', color='red')
plt.plot(X_train, Y_predict, c="red")

plt.title('电磁铁电流与磁场强度的关系')
plt.xlabel('$I$ ($A$)')
plt.ylabel('$B$ ($mT$)')

plt.grid(True)
plt.legend()
plt.show()
```
